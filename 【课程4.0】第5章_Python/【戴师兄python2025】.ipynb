# 函数

```python
import pandas as pd
import os

def concat_data(folder_path, save_path):
    # 存放所有数据表
    all_data = []
    
    # 遍历文件夹中的所有文件
    for file in os.listdir(folder_path):
        if file.endswith(".xlsx"):  # 只处理 Excel 文件
            file_path = os.path.join(folder_path, file)
            df = pd.read_excel(file_path)  # 读取 Excel
            all_data.append(df)  # 加入列表
    
    # 合并所有数据
    merged_df = pd.concat(all_data, ignore_index=True)
    
    # 保存结果
    merged_df.to_excel(rf"{save_path}\合并结果.xlsx", index=False)
    
    print("合并完成！结果已保存到 merged_all.xlsx")
```

```python
# 设置文件夹路径
folder_path = r"E:\教程\【戴师兄 2025 python】\课件\订单数据\2025-09-06"
save_path = r'E:\教程\【戴师兄 2025 python】\课件'
concat_data(folder_path, save_path)
```
