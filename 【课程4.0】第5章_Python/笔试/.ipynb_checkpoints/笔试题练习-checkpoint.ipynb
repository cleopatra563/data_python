| Column Name | Type |
|-------------|------|
| id          | int  |
| salary      | int  |

- 输入: 表格 & n=2
| id | salary |
|-------------|------|
| 1          | 100  |
| 2      | 200  |
| 3      | 300  |
- 输出: 
|getNthHighestSalary(2)|
|--|
|200|

- **示例2**
- 输入: 表格 & n=2
| id | salary |
|-------------|------|
| 1          | 100  |
- 输出:
|getNthHighestSalary(2)|
|--|
|null|
